```python
# Line Plot
# color = color, label = label, linewidth = width of line, alpha = opacity, grid = grid, linestyle = sytle of line
data.ANNUAL.plot(kind = 'line', color = 'g',label = 'ANNUAL', linewidth=1, alpha = 0.5, grid = True,linestyle = ':')
data.YEAR.plot(color = 'r',label = 'YEAR', linewidth=1, alpha = 0.5, grid = True, linestyle = '-.')

plt.legend(loc='upper right')     # legend = puts label into plot
plt.xlabel('x axis')              # label = name of label
plt.ylabel('y axis')
plt.title('Line Plot')            # title = title of plot
plt.show()
```

```python
# Scatter Plot 
# x = attack, y = defense
data.plot(kind='scatter', x='YEAR', y='ANNUAL', alpha = 0.5,color = 'red')
plt.xlabel('YEAR')              # label = name of label
plt.ylabel('ANNUAL')
plt.title('YEAR ANNUAL Scatter Plot')
```

```python
# Histogram
# bins = number of bar in figure
# data.ANNUAL.plot(kind = 'hist',bins = 50,figsize = (12,12))
plt.bar(data['YEAR'], data['ANNUAL'])
plt.show()
```
